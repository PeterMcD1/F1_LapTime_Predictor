```python
import matplotlib.pyplot as plt
import fastf1.plotting
import pandas


fastf1.Cache.enable_cache('../doc_cache')  # replace with your cache directory
import os
os.makedirs('LapAndWeather2019', exist_ok=True)
```

```python
for j in range(1,21):
    session = fastf1.get_session(2019, j, 'Race')
    session.load()
    laps = session.laps
    weather_data = laps.get_weather_data()

    laps = laps.reset_index(drop=True)
    weather_data = weather_data.reset_index(drop=True)
    joined = pandas.concat([laps, weather_data.loc[:, ~(weather_data.columns == 'Time')]], axis=1)
    joined.to_csv('LapAndWeather2019/LapWeather_{}'.format(session.event.EventName))
```
